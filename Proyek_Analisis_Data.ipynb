### Cleaning Data

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana Pengaruh Faktor Musim Pada Jumlah Sewa Sepeda


```python
plt.figure(figsize=(12, 6))
seasonal_counts = hour_df.groupby('season')['cnt'].mean()

plt.bar(x=seasonal_counts.index, height=seasonal_counts, color=['skyblue'])
plt.title('Pengaruh Musim pada Jumlah Sewa Sepeda')
plt.xlabel('Musim (1:Spring, 2:Summer, 3:Fall, 4:Winter)')
plt.ylabel('Rata-rata Jumlah Sewa Sepeda')

plt.xticks(seasonal_counts.index, [str(int(x)) for x in seasonal_counts.index])

plt.show()
```

dapat dilihat bahwa musim gugur memiliki jumlah sewa sepeda tertinggi

### Pertanyaan 2:

```python
workingday_counts = hour_df.groupby('workingday')['cnt'].mean().reset_index()

plt.figure(figsize=(8, 6))
plt.bar(x=workingday_counts['workingday'], height=workingday_counts['cnt'], color=['skyblue', 'lightcoral'])
plt.title('Pengaruh Hari Kerja terhadap Jumlah Sewa Sepeda')
plt.xlabel('Hari Kerja (1: Ya, 0: Tidak)')
plt.ylabel('Rata-rata Jumlah Sewa Sepeda')
plt.xticks([0, 1], ['Bukan Hari Kerja', 'Hari Kerja'])
```

dapat dilihat bahwa pada hari kerja jumlah sewa sepeda lebih banyak di banding bukan hari kerja

### Pertanyaan 3:

```python
holiday_counts = hour_df.groupby('holiday')['cnt'].mean().reset_index()

plt.figure(figsize=(8, 6))
plt.bar(x=holiday_counts['holiday'], height=holiday_counts['cnt'], color=['skyblue', 'lightcoral'])
plt.title('Pengaruh Hari Libur terhadap Jumlah Sewa Sepeda')
plt.xlabel('Hari Libur (1: Ya, 0: Tidak)')
plt.ylabel('Rata-rata Jumlah Sewa Sepeda')
plt.xticks([0, 1], ['Bukan Hari Libur', 'Hari Libur'])

plt.show()
```

dapat dilihat bahwa pada bukan hari libur jumlah sewa sepeda lebih banyak di banding hari libur

## Conclusion

### Conclusion 1
- Musim gugur memiliki jumlah sewa sepeda tertinggi, hal ini dapat karena mungkin dengan kondisi cuaca yang nyaman dan tidak terlalu ekstrem pada musim gugur.
### Conclusion 2
- Sewa sepeda banyak dilakukan pada hari kerja, kemungkinan penyebabnya adalah orang lebih cenderung menggunakan sepeda untuk keperluan transportasi sehari-hari selama hari kerja.
### Conclusion 3
